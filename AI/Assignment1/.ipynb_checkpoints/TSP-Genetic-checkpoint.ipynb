```python
import numpy as np
import random as rd
import time
```

```python
file = open('Sample Tours/reuc_100')
cont = file.read().split('\n')
file.close()
```

```python
eu = (cont[0] == 'euclidean')
n = int(cont[1])
points = []
for i in range(n):
    points.append(list(map(float,cont[i+2].split())))
dis = []
for i in range(n,2*n):
    dis.append(list(map(float,cont[i+2].split())))
```

```python
Q = 10000
P = 100
K = 80
mut = 0.1
num_mut = 40
```

```python
travel_dic = {}
def travel(path):
    global travel_dic,n,dis
    var = tuple(path)
    if travel_dic.get(var,-1) != -1:
        return travel_dic[var]
    ans = 0
    for j in range(n-1):
        ans += dis[path[j]][path[j+1]]
    ans += (dis[path[n-1]][path[0]])
    travel_dic[var] = ans
    return ans

def get_ind(v,fit):
    global P
    l = 0
    r = P
    while (l < r):
        m = (l+r)//2
        if v > fit[m]:
            l = m+1
        else:
            r = m
    return l
```

```python
people = [[] for i in range(P)]
for i in range(P):
    var = [i for i in range(n)]
    while var:
        r = rd.randrange(0,len(var))
        people[i].append(var[r])
        del var[r]
```

```python
fit = [0]*P
start = time.time()
epoch = 0
while time.time() < start + 295:
    epoch+=1
    tot = 0
    for i in range(P):
        fit[i] = Q/travel(people[i])
        tot += fit[i]
    fit[0] = fit[0]/tot
    for i in range(1,P):
        fit[i] = fit[i]/tot
        fit[i] += (fit[i-1])
    select = []
    for i in range(P):
        select.append(people[get_ind(rd.random(),fit)])
    # donot sort select
    offsprings = []
    for i in range(0,P-1,2):
        pos = {}
        for j in range(n):
            pos[select[i][j]] = j
        bel = [-1]*n
        cnt = 0
        for j in range(n):
            if bel[j] == -1:
                cnt += 1
                k = j
                while bel[k] == -1:
                    bel[k] = cnt
                    k = pos[select[i+1][k]]
        c1 = []
        c2 = []
        for j in range(n):
            if bel[j]%2 == 1:
                c1.append(select[i][j])
                c2.append(select[i+1][j])
            else:
                c1.append(select[i+1][j])
                c2.append(select[i][j])
        offsprings.append(c1)
        offsprings.append(c2)
    people = sorted(people,key = travel)
    offsprings = sorted(offsprings, key = travel)
    for i in range(K):
        people[P-i-1] = offsprings[i]
    people = sorted(people,key = travel)
    if (epoch)%200 == 0:
        print(travel(people[0]),flush=True)
#         print(people[0],flush=True)
    j = P-1
    for i in range(num_mut):
        if rd.random() <= mut:
            var = [i for i in range(n)]
            mutant = []
            while var:
                r = rd.randrange(0,len(var))
                mutant.append(var[r])
                del var[r]
            people[j] = mutant
            j-=1
```
