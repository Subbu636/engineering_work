```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('product_data.csv')
del df['timestamp']
df.head()
```

```python
# density of rating
df.describe()
```

```python
# top 10 users by rating
df.groupby(['userId']).mean().sort_values('rating',ascending = False)[:10]
```

```python
# count of users for each product
```

```python
recomm = df.groupby(['productId']).count().sort_values('userId',ascending = False).copy(deep=True)
recomm['num_users'] = recomm['userId']
del recomm['userId']; del recomm['rating']
recomm['rank'] = [i for i in range(1,len(recomm)+1)]
recomm
```

```python
recomm[:5]
```

```python
# non personalized recomandation model based on avg rating of product
np_rec = df.groupby(['productId']).mean().sort_values('rating',ascending = False).copy(deep=True)
np_rec['rank'] = [i for i in range(1,len(np_rec)+1)]
del np_rec['rating']
np_rec
```
