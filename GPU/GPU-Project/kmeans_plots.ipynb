```python
import matplotlib.pyplot as plt
import os
```

```python
file1 = open("ClusterLibs"+os.sep+"N3000K20.txt","r");
file2 = open("ClusterLibs"+os.sep+"N5000K15DENSE.txt","r");
file3 = open("ClusterLibs"+os.sep+"N5000K15SPARSE.txt","r");
file4 = open("ClusterLibs"+os.sep+"N7500K50.txt","r");
file5 = open("ClusterLibs"+os.sep+"N100000K100GRID.txt","r");
file6 = open("ClusterLibs"+os.sep+"N100000K100RD.txt","r");
```

```python
times = []
# 3D lists (dataset, k ,(0,1,2,3,4))
```

```python

x = []
y = []
lines = file1.read().split("\n")
for line in lines:
    arr = line.split()
    if(len(arr)==0):
        break
    x.append(float(arr[0]))
    y.append(float(arr[1]))

ks = [ "16" , "32" , "64" , "128"]
cp_times = []
for k in ks:
    out = open("outputs/"+"output1_"+str(k)+".txt","r");
    lines = out.read().split("\n")
    j = 0
    meansx = []
    meansy = []
    cp = []
    for i in range(5):
        time , converge , dist_time = lines[j].split(" ")
        cp.append(time)
        j+=1
        meansx = []
        meansy = []
        for l in range(int(k)):
            a,b = lines[j].split(" ")
            meansx.append(float(a))
            meansy.append(float(b))
            j+=1
        c = lines[j].split(" ")
        j+=1
    cp_times.append(cp)     
    plt.figure()
    ax = plt.gca()
    ax.scatter(x, y,s = 10)
    ax.scatter(meansx,meansy,marker='x')
    plt.show()
times.append(cp_times)
```

```python
Dataset_names = ['N3000K20','N5000K15DENSE','N5000K15SPARSE','N7500K50','N100000K100GRID','N100000K100RD']
ks = [16,32,64,128]
for j in range(6):
    a = []
    b = []
    c = []
    d = []
    e = []
    for i in range(4):
        a.append(times[j][i][0])
        b.append(times[j][i][1])
        c.append(times[j][i][2])
        d.append(times[j][i][3])
        e.append(times[j][i][4])
    plt.figure()
    plt.plot(ks,a)
    plt.plot(ks,b)
    plt.plot(ks,c)
    plt.plot(ks,d)
    plt.plot(ks,e)
    # plt.plot(ks,f)
    # ax.set_xticks([0,1,2,3,4])
    # ax.set_xticklabels(['3k','5k','7.5k','50k','100k'])
    plt.legend(['standard_cpu','standard_gpu','reinforced_cpu','reinforced_gpu','eff_reinforced_gpu'])
    plt.title('Equivalent Average Labelling Time comparision for '+Dataset_names[j])
    plt.show()
```
