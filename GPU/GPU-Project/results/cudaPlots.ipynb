```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n3k = np.loadtxt('out_cpu_N3000K20 copy.txt')
n5k = np.loadtxt('out_cpu_N5000K15DENSE copy.txt')
p3k = np.loadtxt('N3000K20.txt')
p5k = np.loadtxt('N5000K15DENSE.txt')
c3k = np.array([np.argmax(i) for i in n3k])
c5k = np.array([np.argmax(i) for i in n5k])
```

```python
k3 = 5;k5 = 5
for i in range(k3):
    pc3x = []; pc3y = []
    for j in range(len(c3k)):
        if c3k[j] == i:
            pc3x.append(p3k[j][0])
            pc3y.append(p3k[j][1])
    plt.scatter(pc3x,pc3y,marker='.')
plt.title('3k points')
plt.show()
for i in range(k5):
    pc5x = []; pc5y = []
    for j in range(len(c5k)):
        if c5k[j] == i:
            pc5x.append(p5k[j][0])
            pc5y.append(p5k[j][1])
    plt.scatter(pc5x,pc5y,marker='.')
plt.title('5k points')
plt.show()
```
