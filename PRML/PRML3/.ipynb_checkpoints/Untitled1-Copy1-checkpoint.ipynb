```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from scipy.cluster.vq import kmeans as kmeanss
from scipy.cluster.vq import vq as vqs

from math import sqrt
import math
from matplotlib.mlab import PCA
import numpy as np
import numpy.linalg as linalg
from random import sample
import csv
import copy
import os
import glob
```

```python
def initializecentroids(data,k):
    rand_ids = sample(range(data.shape[0]),k)
    centroids = []
    for i in rand_ids:
        d = data.loc[i,:]
        centroids.append(d)
    #print(centroids)
    return centroids

```

```python
def euclid_dist(x,y):
    n=len(x)
    s=0
    for i in range(n):
        sq = (x[i]-y[i])**2
        s +=sq
    return sqrt(s)
```

```python
def assigncluster(data,k,centroids):
    acluster=[]
    for i in range(data.shape[0]):
        minind=0
        m=1000000
        for j in range(len(centroids)):
            a = euclid_dist(data.loc[i,:],centroids[j])
            if a <= m:
                m=a
                minind=j
        acluster.append(minind)
    return acluster           
```

```python
def meaneuclid(datalist):
    d=[]
    for i in range(len(datalist[0])):
        m=0
        for j in range(len(datalist)):
            m += datalist[j][i]
        d.append(m/(len(datalist)))
        
    return d
```

```python
def estimcentroid(data,k,assclusters):
    centroids = []
    dp=[]
    for i in range(k):
        dp=[]
        for j in range(len(assclusters)):
            if assclusters[j]==i:
                dp.append(data.loc[j,:])
        #print(dp)
        #print(centroids)
        centroids.append(meaneuclid(dp))
    return centroids
```

```python
def kmeans(data,k):
    centroids = initializecentroids(data,k)
    iter_num=1
    MAX=30
    while iter_num <= MAX:
        assignedclusters = assigncluster(data,k,centroids)
        #print(assignedclusters)
        oldcentroids = copy.deepcopy(centroids)
        centroids = estimcentroid(data,k,assignedclusters)
        iter_num+=1
   # print (assignedclusters)
    return assignedclusters,centroids      
```

```python
def counter(data,k):
    #kmeans
    cen,asstates = kmeanss(pd.DataFrame(data),k)
    idx,_ = vqs(pd.DataFrame(data),cen)
    #number of states = k
    totc = np.zeros(k)
    transc = np.zeros((k,k))
    for i in range(np.size(data,0)):
        a = asstates[i]
        totc[a] = totc[a]+1
    
    for j in range(np.size(data,0) - 1):
        a = asstates[j]
        b = asstates[j+1]
        transc[a][b] = transc[a][b]+1     
    return totc,transc,idx
        
```

```python
def dist(a,b):
    n = np.size(a)
    d =0
    #print(n)
    if n!=1:
        for i in range(n):
            d = d + (a[i]-b[i])**2
    else:
        d = (a-b)**2
    return sqrt(d)


def dtw(series1,series2):
    m = len (series1)
    n = len (series2)
    Dtw = np.zeros((n,m))
    #row
    for i in range(1,m):
        Dtw[0,i] = dist(series1[i],series2[0]) + Dtw[0,i-1]
    for i in range(1,n):
        Dtw[i,0] = dist(series1[0],series2[i]) + Dtw[i-1,0]
    for i in range(1,n):
        for j in range(1,m):
            c = dist(series1[j],series2[i])
            Dtw[i,j] = c + np.min([Dtw[i-1,j],Dtw[i,j-1],Dtw[i-1,j-1]])
    return Dtw[n-1,m-1]
```
